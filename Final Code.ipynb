```python
import tkinter as tk
from tkinter import messagebox

# Sample questions and answers
questions = [
    {"question": "What is the capital of France?", "options": ["Paris", "London", "Berlin", "Rome"], "answer": "Paris"},
    {"question": "What is 5 + 7?", "options": ["10", "12", "13", "11"], "answer": "12"},
    {"question": "Which planet is known as the Red Planet?", "options": ["Earth", "Mars", "Jupiter", "Venus"], "answer": "Mars"}
]

class QuizGame:
    def __init__(self, root):
        self.root = root
        self.root.title("Quiz Game with XP")
        self.xp = 0
        self.q_index = 0

        self.question_label = tk.Label(root, text="", font=("Arial", 16), wraplength=400)
        self.question_label.pack(pady=20)

        self.buttons = []
        for i in range(4):
            btn = tk.Button(root, text="", font=("Arial", 14), width=20, command=lambda i=i: self.check_answer(i))
            btn.pack(pady=5)
            self.buttons.append(btn)

        self.xp_label = tk.Label(root, text="XP: 0", font=("Arial", 14))
        self.xp_label.pack(pady=10)

        self.load_question()

    def load_question(self):
        if self.q_index < len(questions):
            q = questions[self.q_index]
            self.question_label.config(text=q["question"])
            for i, option in enumerate(q["options"]):
                self.buttons[i].config(text=option)
        else:
            messagebox.showinfo("Quiz Complete", f"You finished the quiz with {self.xp} XP!")
            self.root.quit()

    def check_answer(self, selected_index):
        correct = questions[self.q_index]["answer"]
        chosen = questions[self.q_index]["options"][selected_index]
        if chosen == correct:
            self.xp += 10
            self.xp_label.config(text=f"XP: {self.xp}")
        self.q_index += 1
        self.load_question()

# Run the game
root = tk.Tk()
game = QuizGame(root)
root.mainloop()
```
